```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import allantools
```

```python
pd.set_option("display.precision", 15)      # Show up to 15 decimal places
#baseName = 'fixedL1l52'
baseName = 'fixedL1l52'
dirName  = 'labData/'
rowLimit = 2000 # Rows to keep after joining ticc and timTp samples
```

```python
timTp = pd.read_csv(f"{dirName}/{baseName}.timTpB.csv")
timTp['fn'] = baseName
```

```python
timTp
```

```python
qeRange = timTp.qErr.max()-timTp.qErr.min()
print(f"qErr range implies TCXO clock of {1e6/qeRange} MHz")
```

```python
beg = 2000
end = 3000
study = timTp.qErr[beg:end]
study = study.reset_index(drop=True)
```

```python
# Good inverted hanging bridge in baseline1
beg = 10015
end = 10500
study = timTp.qErr[beg:end]
study = study.reset_index(drop=True)
```

```python
wraps = np.where(np.abs(np.diff(study)) > 0.55*qeRange)[0]  # Threshold ~70% of full scale
wrapPeriods = np.diff(wraps)
```

```python
plt.figure(figsize=(22, 6))
plt.title(f"qErr Over {end-beg} Samples")
plt.xlabel("Elapsed Time (s)")
plt.ylabel("qErr (ps)")
plt.vlines(x=wraps, ymin=-4101.5, ymax=4101.5, color='red', linestyle='--', label='wraps')
#ax2 = plt.gca().twinx()
#ax2.plot(deltas, label='difference', color='red', marker='.')
plt.plot(study, 'o-')
```

```python
plt.figure(figsize=(22, 5))
plt.plot(np.unwrap(study, period=qeRange))
plt.xlabel("Elapsed Time (s)")
plt.ylabel("Cumulative qErr (ps)")
plt.title(f"Cumulative qErr Over {end-beg} Samples")
```
